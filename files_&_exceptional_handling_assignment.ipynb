```python
#6 Write a python program that logs and error message to a log file when division by zero exception occurs.

import logging
logging.basicConfig(filename = "program.log", level= logging.DEBUG)
try:
  10/0
except ZeroDivisionError as e:
  logging.error("I am handling zero division error", e)
```
